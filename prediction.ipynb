```python
import tensorflow as tf
```

```python
import pickle
from tensorflow.keras.models import load_model
import pandas as pd
import numpy as np
```

```python
##Load trained model,scalar,pickle,onehot
models=load_model('models.h5')
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)
with open('one_hot_encoder_geo.pkl','rb') as file:
    one_hot_encoder_geo=pickle.load(file)
with open('scalar.pkl','rb') as file:
    scalar=pickle.load(file)
```

```python
##input example
input_data={
    'CreditScore':600,
    'Geography':'Spain',
    'Gender':'Male',
    'Age':42,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geo_encoded=one_hot_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
df_new_data=pd.DataFrame([input_data])
df_new_data
```

```python
df_new_data['Gender']=label_encoder_gender.transform(df_new_data['Gender'])
df_new_data
```

```python
#concating
df_new_data=pd.concat([df_new_data.drop(columns='Geography',axis=1) ,geo_encoded_df],axis=1)
```

```python
df_new_data
```

```python
input_scaled=scalar.transform(df_new_data)
```

```python
input_scaled
```

```python
prediction=models.predict(input_scaled)
```

```python
prediction
```

```python
if(prediction>0.5):
    print("The customer is likely to churn")
else:
    print("The customer is not likely to churn")
```
